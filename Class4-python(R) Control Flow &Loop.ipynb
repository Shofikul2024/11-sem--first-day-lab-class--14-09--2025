```python
'''    

Code 1: ‡¶ó‡ßç‡¶∞‡ßá‡¶° ‡¶∏‡¶ø‡¶∏‡ßç‡¶ü‡ßá‡¶Æ (Grade System)
marks = 66
if marks >= 80 and marks <= 100:
    print("A+")

elif marks >= 70 and marks < 80:
    print("A")

elif marks >= 60 and marks < 70:
    print("A-")

elif marks >= 50 and marks < 60:
    print("B")

elif marks >= 40 and marks < 50:
    print("C")

elif marks >= 33 and marks < 40:
    print("D")

else:
    print("F")


‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ ‡¶¨‡ßç‡¶Ø‡¶æ‡¶ñ‡ßç‡¶Ø‡¶æ:
‡¶è‡¶ñ‡¶æ‡¶®‡ßá ‡¶è‡¶ï‡¶ü‡¶ø ‡¶õ‡¶æ‡¶§‡ßç‡¶∞‡ßá‡¶∞ ‡¶®‡¶Æ‡ßç‡¶¨‡¶∞ marks ‡¶è‡¶∞ ‡¶â‡¶™‡¶∞ ‡¶≠‡¶ø‡¶§‡ßç‡¶§‡¶ø ‡¶ï‡¶∞‡ßá ‡¶ó‡ßç‡¶∞‡ßá‡¶° ‡¶®‡¶ø‡¶∞‡ßç‡¶ß‡¶æ‡¶∞‡¶£ ‡¶ï‡¶∞‡¶æ ‡¶π‡¶ö‡ßç‡¶õ‡ßá‡•§ ‡¶Ø‡¶¶‡¶ø ‡¶®‡¶Æ‡ßç‡¶¨‡¶∞ ‡ßÆ‡ß¶‚Äì‡ßß‡ß¶‡ß¶ ‡¶è‡¶∞ ‡¶Æ‡¶ß‡ßç‡¶Ø‡ßá ‡¶π‡ßü ‡¶§‡¶¨‡ßá A+, ‡ß≠‡ß¶‚Äì‡ß≠‡ßØ ‡¶π‡¶≤‡ßá A, ‡ß¨‡ß¶‚Äì‡ß¨‡ßØ ‡¶π‡¶≤‡ßá A-, ‡¶á‡¶§‡ßç‡¶Ø‡¶æ‡¶¶‡¶ø‡•§ ‡ß©‡ß© ‡¶è‡¶∞ ‡¶ï‡¶Æ ‡¶π‡¶≤‡ßá F ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶π‡¶¨‡ßá‡•§

English Explanation:
This code checks a student's marks and prints their grade. If the marks are between 80‚Äì100, it prints A+. If between 70‚Äì79, it prints A, 60‚Äì69 ‚Üí A-, and so on. If marks are below 33, it prints F.

üîπ Code 2: ‡¶¨‡ßü‡¶∏ ‡¶ì ‡¶™‡¶æ‡¶∞‡¶Æ‡¶ø‡¶∂‡¶® ‡¶ö‡ßá‡¶ï (Age & Permission Check)
age = 30
permission = False

if age >= 20:
    if permission == True:
        print("You can join with us")
    else:
        print("You can't join with us")
else:
    print("You are under aged")


‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ ‡¶¨‡ßç‡¶Ø‡¶æ‡¶ñ‡ßç‡¶Ø‡¶æ:
‡¶è‡¶ñ‡¶æ‡¶®‡ßá ‡¶™‡ßç‡¶∞‡¶•‡¶Æ‡ßá ‡¶¨‡ßü‡¶∏ ‡¶¶‡ßá‡¶ñ‡¶æ ‡¶π‡¶ö‡ßç‡¶õ‡ßá‡•§ ‡¶Ø‡¶¶‡¶ø ‡¶¨‡ßü‡¶∏ ‡ß®‡ß¶ ‡¶¨‡¶æ ‡¶§‡¶æ‡¶∞ ‡¶¨‡ßá‡¶∂‡¶ø ‡¶π‡ßü ‡¶è‡¶¨‡¶Ç permission ‡¶∏‡¶§‡ßç‡¶Ø‡¶ø (True) ‡¶π‡ßü, ‡¶§‡¶¨‡ßá ‡¶∏‡ßá ‡¶Ø‡ßã‡¶ó ‡¶¶‡¶ø‡¶§‡ßá ‡¶™‡¶æ‡¶∞‡¶¨‡ßá‡•§ ‡¶Ø‡¶¶‡¶ø ‡¶¨‡ßü‡¶∏ ‡¶†‡¶ø‡¶ï ‡¶•‡¶æ‡¶ï‡ßá ‡¶ï‡¶ø‡¶®‡ßç‡¶§‡ßÅ ‡¶Ö‡¶®‡ßÅ‡¶Æ‡¶§‡¶ø ‡¶®‡¶æ ‡¶•‡¶æ‡¶ï‡ßá, ‡¶§‡¶¨‡ßá ‡¶Ø‡ßã‡¶ó ‡¶¶‡¶ø‡¶§‡ßá ‡¶™‡¶æ‡¶∞‡¶¨‡ßá ‡¶®‡¶æ‡•§ ‡¶Ü‡¶∞ ‡¶¨‡ßü‡¶∏ ‡ß®‡ß¶ ‡¶è‡¶∞ ‡¶ï‡¶Æ ‡¶π‡¶≤‡ßá "under aged" ‡¶¶‡ßá‡¶ñ‡¶æ‡¶¨‡ßá‡•§

English Explanation:
This code checks if a person is 20 or older and has permission. If both conditions are true, they can join. If the age is enough but permission is False, they cannot join. If the age is below 20, it prints "You are under aged".

üîπ Code 3: for ‡¶≤‡ßÅ‡¶™ (For Loop)
girls = ["Katrina", "Carina", "Jarina", "Sakhina"]
for gf in girls:
    print(gf)


‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ ‡¶¨‡ßç‡¶Ø‡¶æ‡¶ñ‡ßç‡¶Ø‡¶æ:
‡¶è‡¶ñ‡¶æ‡¶®‡ßá for loop ‡¶¨‡ßç‡¶Ø‡¶¨‡¶π‡¶æ‡¶∞ ‡¶ï‡¶∞‡ßá ‡¶≤‡¶ø‡¶∏‡ßç‡¶ü‡ßá‡¶∞ ‡¶™‡ßç‡¶∞‡¶§‡¶ø‡¶ü‡¶ø ‡¶®‡¶æ‡¶Æ ‡¶è‡¶ï‡ßá ‡¶è‡¶ï‡ßá ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶ï‡¶∞‡¶æ ‡¶π‡¶ö‡ßç‡¶õ‡ßá‡•§

English Explanation:
Here, a for loop is used to print each name from the list one by one.

üîπ Code 4: while ‡¶≤‡ßÅ‡¶™ (While Loop)
girls = ["Katrina", "Carina", "Jarina", "Sakhina"]
chokkor = 0
while chokkor < len(girls):
    print(girls[chokkor])
    chokkor = chokkor + 1


‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ ‡¶¨‡ßç‡¶Ø‡¶æ‡¶ñ‡ßç‡¶Ø‡¶æ:
while loop ‡¶¨‡ßç‡¶Ø‡¶¨‡¶π‡¶æ‡¶∞ ‡¶ï‡¶∞‡ßá ‡¶≤‡¶ø‡¶∏‡ßç‡¶ü‡ßá‡¶∞ ‡¶®‡¶æ‡¶Æ‡¶ó‡ßÅ‡¶≤‡ßã ‡¶è‡¶ï‡ßá ‡¶è‡¶ï‡ßá ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶ï‡¶∞‡¶æ ‡¶π‡¶ö‡ßç‡¶õ‡ßá‡•§ ‡¶è‡¶ñ‡¶æ‡¶®‡ßá ‡¶ï‡¶æ‡¶â‡¶®‡ßç‡¶ü‡¶æ‡¶∞ chokkor ‡¶™‡ßç‡¶∞‡¶§‡¶ø ‡¶ß‡¶æ‡¶™‡ßá ‡ßß ‡¶ï‡¶∞‡ßá ‡¶¨‡¶æ‡ßú‡¶õ‡ßá‡•§

English Explanation:
Using a while loop, each name from the list is printed. The counter chokkor increases by 1 each time until it reaches the list length.

üîπ Code 5: break ‡¶∏‡ßç‡¶ü‡ßá‡¶ü‡¶Æ‡ßá‡¶®‡ßç‡¶ü (Break Statement)
girls = ["Katrina", "Carina", "Jarina", "Sakhina"]
for gf in girls:
    if gf == "Jarina":
        break
    print(gf)


‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ ‡¶¨‡ßç‡¶Ø‡¶æ‡¶ñ‡ßç‡¶Ø‡¶æ:
‡¶Ø‡¶ñ‡¶® ‡¶≤‡ßÅ‡¶™‡ßá Jarina ‡¶™‡¶æ‡¶ì‡ßü‡¶æ ‡¶Ø‡¶æ‡¶¨‡ßá ‡¶§‡¶ñ‡¶® ‡¶≤‡ßÅ‡¶™ ‡¶¨‡¶®‡ßç‡¶ß ‡¶π‡ßü‡ßá ‡¶Ø‡¶æ‡¶¨‡ßá (break)‡•§ ‡¶§‡¶æ‡¶á ‡¶è‡¶∞ ‡¶™‡¶∞‡ßá ‡¶Ü‡¶∞ ‡¶ï‡¶ø‡¶õ‡ßÅ ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶π‡¶¨‡ßá ‡¶®‡¶æ‡•§

English Explanation:
When the loop finds "Jarina", it breaks and stops execution. So, names after "Jarina" are not printed.

üîπ Code 6: continue ‡¶∏‡ßç‡¶ü‡ßá‡¶ü‡¶Æ‡ßá‡¶®‡ßç‡¶ü (Continue Statement)
girls = ["Katrina", "Carina", "Jarina", "Sakhina"]
for gf in girls:
    if gf == "Carina":
        continue
    print(gf)


‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ ‡¶¨‡ßç‡¶Ø‡¶æ‡¶ñ‡ßç‡¶Ø‡¶æ:
‡¶Ø‡¶ñ‡¶® "Carina" ‡¶™‡¶æ‡¶ì‡ßü‡¶æ ‡¶Ø‡¶æ‡¶¨‡ßá ‡¶§‡¶ñ‡¶® ‡¶∏‡ßá‡¶ü‡¶æ ‡¶¨‡¶æ‡¶¶ ‡¶¶‡¶ø‡ßü‡ßá ‡¶≤‡ßÅ‡¶™ ‡¶ö‡¶æ‡¶≤‡¶ø‡ßü‡ßá ‡¶Ø‡¶æ‡¶¨‡ßá‡•§ ‡¶§‡¶æ‡¶á "Carina" ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶π‡¶¨‡ßá ‡¶®‡¶æ‡•§

English Explanation:
When "Carina" is found, the loop skips it and continues. So, "Carina" is not printed.

üîπ Code 7: ‡¶≤‡¶ú‡¶ø‡¶ï‡ßç‡¶Ø‡¶æ‡¶≤ ‡¶Ö‡¶™‡¶æ‡¶∞‡ßá‡¶ü‡¶∞ (Logical Operators)
age = 19
permission = True

if age >= 18 and permission == True:
    print("Welcome")


‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ: ‡¶â‡¶≠‡ßü ‡¶∂‡¶∞‡ßç‡¶§ (‡¶¨‡ßü‡¶∏ ‚â•‡ßß‡ßÆ ‡¶è‡¶¨‡¶Ç permission=True) ‡¶™‡ßÇ‡¶∞‡¶£ ‡¶π‡¶≤‡ßá Welcome ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶π‡¶¨‡ßá‡•§
English: Prints Welcome only if both conditions are true.

age = 10
permission = True

if age >= 18 or permission == True:
    print("Welcome")


‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ: ‡¶Ö‡¶®‡ßç‡¶§‡¶§ ‡¶è‡¶ï‡¶ü‡¶ø ‡¶∂‡¶∞‡ßç‡¶§ ‡¶∏‡¶§‡ßç‡¶Ø‡¶ø ‡¶π‡¶≤‡ßá Welcome ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶π‡¶¨‡ßá‡•§ ‡¶è‡¶ñ‡¶æ‡¶®‡ßá ‡¶¨‡ßü‡¶∏ ‡¶õ‡ßã‡¶ü ‡¶π‡¶≤‡ßá‡¶ì permission=True, ‡¶§‡¶æ‡¶á Welcome ‡¶π‡¶¨‡ßá‡•§
English: Prints Welcome if at least one condition is true. Here, permission is True, so it prints.

age = 10
permission = True

if not age > 18:
    print("Not Welcome")


‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ: ‡¶Ø‡¶¶‡¶ø ‡¶¨‡ßü‡¶∏ ‡ßß‡ßÆ ‡¶è‡¶∞ ‡¶¨‡ßá‡¶∂‡¶ø ‡¶®‡¶æ ‡¶π‡ßü (‡¶Æ‡¶æ‡¶®‡ßá ‚â§‡ßß‡ßÆ), ‡¶§‡¶¨‡ßá Not Welcome ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶π‡¶¨‡ßá‡•§
English: Prints Not Welcome if age is not greater than 18.

üîπ Code 8: Nested Loop (‡¶®‡ßá‡¶∏‡ßç‡¶ü‡ßá‡¶° ‡¶≤‡ßÅ‡¶™)
girls = ["Katrina", "Carina", "Jarina"] 
girls_papa = ["KatrinaBap", "CarinaBap", "JarinaBap", "SakhinaBap"] 

for gf in girls:
    print(gf)
    for papa in girls_papa:
        print(papa)


‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ ‡¶¨‡ßç‡¶Ø‡¶æ‡¶ñ‡ßç‡¶Ø‡¶æ:
‡¶™‡ßç‡¶∞‡¶•‡¶Æ‡ßá ‡¶è‡¶ï‡¶ü‡¶ø girl ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶π‡¶¨‡ßá, ‡¶§‡¶æ‡¶∞‡¶™‡¶∞ ‡¶§‡¶æ‡¶∞ ‡¶∏‡¶æ‡¶•‡ßá ‡¶≤‡¶ø‡¶∏‡ßç‡¶ü‡ßá‡¶∞ ‡¶∏‡¶¨ papa ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶π‡¶¨‡ßá‡•§ ‡¶è‡¶≠‡¶æ‡¶¨‡ßá ‡¶™‡ßç‡¶∞‡¶§‡¶ø‡¶ü‡¶ø ‡¶Æ‡ßá‡ßü‡ßá‡¶∞ ‡¶ú‡¶®‡ßç‡¶Ø ‡¶∏‡¶¨ ‡¶¨‡¶æ‡¶¨‡¶æ‡¶∞ ‡¶®‡¶æ‡¶Æ ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶π‡¶¨‡ßá‡•§

English Explanation:
First, one girl is printed, then all papa names are printed for that girl. This repeats for each girl.

üîπ Code 9: ‡¶π‡ßç‡¶Ø‡¶æ‡¶≤‡ßã ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü (Hello Print)
print('hello')


‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ ‡¶¨‡ßç‡¶Ø‡¶æ‡¶ñ‡ßç‡¶Ø‡¶æ:
‡¶è‡¶ü‡¶ø ‡¶∂‡ßÅ‡¶ß‡ßÅ hello ‡¶™‡ßç‡¶∞‡¶ø‡¶®‡ßç‡¶ü ‡¶ï‡¶∞‡ßá‡•§

English Explanation:
This simply prints hello.

'''
```
